```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
# Insert Node
    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
# Print the Tree
    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print( self.data),
        if self.right:
            self.right.PrintTree()
            
# =============================== 3.1 ======================================================================== 
# Left -> Root -> Right
    def inorderTraversal(self, root):
        res = []
        if root:
            res = self.inorderTraversal(root.left)
            res.append(root.data)
            res = res + self.inorderTraversal(root.right)
        return res
# Root -> Left -> Right 
    def preorderTraversal(self, root):
        res = []
        if root:
            res.append(root.data)
            res = res + self.preorderTraversal(root.left)
            res = res + self.preorderTraversal(root.right)
        return res
# Left -> Right -> Root
    def postorderTraversal(self, root):
            res = []
            if root:
                res = res + self.postorderTraversal(root.left)
                res = res + self.postorderTraversal(root.right)
                res.append(root.data)
            return res
# =============================== 3.2 ========================================================================            
    def count_nodes(self,root):
        if root is None:
            return 0
        else:
            return 1 + self.count_nodes(root.left) + self.count_nodes(root.right)

    def hieght(self,root):
        if root is None:
            return -1
        else:

            left = self.hieght(root.left)
            right = self.hieght(root.right)

            if left > right :
                return left + 1
            else:
                return right + 1
        
        
        
        
# root = Node(20)
# root.insert(12)
# root.insert(35)
# root.insert(5)
# root.insert(15)
# root.insert(25)
# root.insert(40)
# root.insert(30)
# root.insert(13)
# """
#             20
#            /   \  
#          12    35
#         /  \   /  \ 
#        5   15 25  40
#           /     \  
#          13      30
        
# """
"""
               27
            /     \ 
           14      35
          /  \    /  \ 
         10  19  31   42 

"""
root = Node(27)
root.insert(14)
root.insert(35)
root.insert(10)
root.insert(19)
root.insert(31)
root.insert(42)
print("=========== Preorder ==========")
print(root.preorderTraversal(root)) 
print("========== Inorder ==========")
print(root.inorderTraversal(root)) 
print("========== Postorder ==========")
print(root.postorderTraversal(root)) 
print("========== Total_nodes ==========")
print(root.count_nodes(root))
print("========== Height_of_Tree ==========")
print(root.hieght(root))
```

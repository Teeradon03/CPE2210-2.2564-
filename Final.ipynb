```python
import sys

class Graph():
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)]
                      for row in range(vertices)]

    def printSolution(self, dist, minpath):
        ch = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
        print("Vertex \t Distance from Source \t Minimum Path")
        for node in range(self.V):
            pass
        print(ch[node], "\t\t", dist[node], "\t\t", minpath[node])

    def minDistance(self, dist, sptSet):
        min = sys.maxsize
        for v in range(self.V):
            if dist[v] < min and sptSet[v] == False:
                min = dist[v]
                min_index = v
        return min_index

    def dijkstra(self, src):
        dist = [sys.maxsize] * self.V
        dist[src] = 0
        sptSet = [False] * self.V
        minpath = [""] * self.V
        ch = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
        for cout in range(self.V):
            u = self.minDistance(dist, sptSet)
            sptSet[u] = True
            for v in range(self.V):
                if self.graph[u][v] > 0 and sptSet[v] == False and \
                        dist[v] > dist[u] + self.graph[u][v]:
                    dist[v] = dist[u] + self.graph[u][v]
                    minpath[v] = minpath[u] + " -> " + str(ch[u])
        self.printSolution(dist, minpath)


g = Graph(7) 
g.graph = [
 #   a  b  c  d  e  f  g 
    [0, 3, 0, 6, 1, 0, 0],  # a
    [3, 0, 6, 0, 5, 0, 0],  # b
    [0, 6, 0, 0, 3, 10, 0],  # c
    [6, 0, 0, 0, 6, 0, 7],  # d
    [1, 5, 3, 6, 0, 0, 4],  # e
    [0, 0, 10, 0, 0, 0, 0],  # f
    [0, 0, 0, 7, 4, 0, 0],  # g
]
g.dijkstra(0)

```

```python
##   Binary tree!!!! 
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

    def preorder(self):
        print(self.val, end=' ')
        if self.left:
            self.left.preorder()
        if self.right:
            self.right.preorder()
    def inorder(self):
        if self.left:
            self.left.inorder()
        print(self.val, end=' ')
        if self.right:
            self.right.inorder()
    def postorder(self):
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.val, end=' ')
## ---------------------3.1------------------
    def size_of_tree(self,node):
        if node is None:
            return 0
        else:
            return (1 + self.size_of_tree(node.left) + self.size_of_tree(node.right))

## ------------------ 3.2 -------------
    def height(self,node):
        if node is None:
            return -1
        else:
            
            left_h = self.height(node.left)
            right_h = self.height(node.right)

            if left_h > right_h:
                return left_h + 1
            else:
                return right_h +1

 
"""
              200 
            /     \ 
         300       400
        /  \      /  \ 
      500  600  700   800

"""


root = Node(200)
root.left = Node(300)
root.right = Node(400)
root.left.left = Node(500)
root.left.right = Node(600)
root.right.left = Node(700)
root.right.right = Node(800)

print("Preorder: ", end=" ")
root.preorder()
print("\nInorder: ", end=" ")
root.inorder()
print("\nPostorder:", end=" ")
root.postorder()
print(f"\ntotal_node_tree : {root.size_of_tree(root)}")
print(f"Height_tree : {root.height(root)}")
```
